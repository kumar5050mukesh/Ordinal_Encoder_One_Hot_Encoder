```python
# Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in
# a machine learning project.
"""Targeted guided ordinal encoding works if the column has multiple entried and this column has strong correlation 
with other column 

Here is an example of the Target Guided Ordinal encoding
"""

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
df=pd.DataFrame({"city":["gaya","patna","gaya","patna","bhagalpur","muzzafarpur","aurangabad","sasaram","sherghati"],
                 "rent":[200,456,345,234,456,456,789,345,567]})
df2=df.groupby(df["city"])["rent"].mean().to_dict()
df["city_encoded"]=df["city"].map(df2)
df


"""the prime condition for the Targeted Guided Ordinal Encoding is to have catagorical column with the repeating names 
and the strong realtion of this column with the other column"""

```

```python
# Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?
"""Covariance is a important step is in statistical analysis , the covariance determines how the independent 
variables are linked. If the value of the covariance comes out to be positive then if the value of x increases then value of 
y also increases if the value of x decreases then value of y also decreases

If the value of the covariance comes out to be neagative then in this case if the value of x increased then the value of
y decreases and if the value of x decreases then value of y increases

coavraince is calcualted using the formuala 

Cov(X,Y) = Σ [(Xi - μX) * (Yi - μY)] / (n - 1)
Here is the Python code for the implenatation of covariance """

import numpy as np
import pandas as pd
df=pd.DataFrame({"one":[34,56,78,45,23,45],
              "two":[45,23,45,67,23,45]})
import numpy as np

cov_mat = np.cov([df['one'], df['two']])
cov_mat
"""this result of the cov_mat shows that the covaraince within  the column one is 350.96 and the covaraince is 56.46 and 
the covariance within the column two is 274.26 and the covaraince between the column is 56.26"""



```

```python
# Q5. Calculate the covariance matrix for the following variables in a dataset: Age, Income, and Education
# level. Interpret the results.
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({"age": [22, 34, 45, 67, 34, 56, 23],
                   "income": [22000, 34000, 45000, 56000, 78000, 32000, 88000],
                   "education": ["matric", "ia", "ba", "matric", "ia", "ba", "ba"]})

encoder = OrdinalEncoder(categories=[["matric", "ia", "ba"]])
encoded = encoder.fit_transform(df[["education"]])
df_encoded = pd.DataFrame(encoded, columns=["encoded_education"])

df_final = pd.concat([df, df_encoded], axis=1)

df_matrix = df_final[["age", "income", "encoded_education"]].cov()
print(df_matrix)


```

```python
# Q7. You are analyzing a dataset with two continuous variables, "Temperature" and "Humidity", and two
# categorical variables, "Weather Condition" (Sunny/Cloudy/Rainy) and "Wind Direction" (North/South/
# East/West). Calculate the covariance between each pair of variables and interpret the results.
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({"temp": [22, 34, 45, 45, 23, 19, 18,],
                   "humidity": [99, 88, 76, 56, 78, 32, 88],
                   "wind_direction": ["north", "south", "north", "east", "west", "north", "south"],
                   "wind_condition": ["sunny", "cloudy", "rainy", "rainy", "sunny", "cloudy", "rainy"]})

encoder = OrdinalEncoder(categories=[["north","south","east","west"],["sunny","cloudy","rainy"]])
encoded = encoder.fit_transform(df[["wind_direction","wind_condition"]])
df_encoded = pd.DataFrame(encoded, columns=["encoded_wind_direction","encoded_wind_condition"])


df_final = pd.concat([df, df_encoded], axis=1)

df_matrix = df_final[["temp", "humidity", "encoded_wind_direction","encoded_wind_condition"]].cov()
print(df_matrix)
```
